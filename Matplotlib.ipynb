```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_title('LARGER PLOT')
axes2.plot(y,x,'b')
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig,axes = plt.subplots(nrows =2, ncols = 1, figsize = (8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x ** 2, label = 'X squared')
axes.plot(x, x ** 3, label = 'X cubed')
axes.legend(loc = 'best')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color = 'green', linewidth = 3, alpha =0.7)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color = 'green', lw = 3, ls = '-.')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color = 'green', lw = 3, ls = '-', marker = 'o', markersize = 8, 
          markerfacecolor = 'purple', markeredgewidth = 3, markeredgecolor = 'black')
```
