```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.plot(x,z)
ax2.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim(left = 20, right = 22)
ax2.set_ylim(bottom = 30, top = 50)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y, linewidth = 3, linestyle = '--')
axes[1].plot(x,z, color = 'red', linewidth = 4)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12,2))
axes[0].plot(x,y, color = 'blue', linewidth = 4)
axes[1].plot(x,z, color = 'red', linewidth = 3, linestyle = '--')
```
